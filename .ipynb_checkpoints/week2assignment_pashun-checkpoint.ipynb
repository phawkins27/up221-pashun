### Finding total count for a variable 
Here the outcome shows us the individal count of the different types of parks present in the data. 

```python
# variable count 
parks['PARK_NAME'].value_counts()
```

The following function save the actual count of parks as its own variable. 
The outcome was the same as before, and it is showing the individual count of parks. 

```python
# saving parks count as a variable
park_count = parks['PARK_NAME'].value_counts()
park_count
```

```python
# checking data type of park
type(park_count)
```

### Converting series into a dataframe
The out come is seeing the variable I created as a dataframe. The data frame is showing both the head and tails of the data. Also, the data frame is now assigned to the variable to "park_count". 

```python
# Converting the series into a dataframe
park_count = park_count.reset_index()
park_count
```

```python
# Checking data type
type(park_count)
```

### Renaming Columns
The output is the reation of a list

```python
# creating current columns as a list
park_count.columns.to_list()
```

This will result in the actuall change of the names in the data frame to "parks" and "count"

```python
park_count.columns = ['park', 'count']
```

This function projects the actual columns in the data frame, so you can see that they were renamed. 

```python
#printing park_count 
park_count
```

### Creating Bar Plot 

The out come is a bargraph showcasing park count 

```python
# Creates a bar plot 
park_count.plot()
```

### Defining arguments in bargraph

The outcome is assigning the unnameed variables in the graph. The graph will now have labled x and y axis and a  title. 

```python
# Defining the variables for the bargraph
park_count.plot.bar(x = 'park', y = 'count', title = 'Number of parks in LA county')
```

### Trimming Data
Here, the number of columns will be decreased, and colmns not needed will be elminated. First to be displayed is the original data before elimination.

```python
# the output is the original data's colmns
parks.info()
```

The outcome is eliminating the undesired columns, and then assigning the trimmed data frame to a new variable. 

```python
# subset of columns in the dataframe 
desired_parks=parks[['PARK_NAME','CENTER_LAT','CENTER_LON','geometry']]
```

Priniting the trimmed data set to see the outcome of the previous function

```python
# Priniting the trimmed data set
desired_parks.head()
```

### Querying and filtering the data

After querying the data the result was only seeing data from the park I specified. 

```python
#filtering data to only see the Atlantic Avenue Park data
desired_parks.query("PARK_NAME == 'Atlantic Avenue Park'")
```

### Plotting
The following function will help visuallize the data based on the dataframe that was trimmed. 

```python
#ploting trimmed data of parks
desired_parks.plot()
```

```python
# provides a look at the geometry column 
desired_parks.geometry
```

### Adding legend, fixing size, and providing color to plot
The outcome will be the same plot but with more detail. There will be 

```python
desired_parks.plot(
            figsize=(100,60),   # plot size
            column = 'PARK_NAME',  
            legend = True,     # adds a legend           
            legend_kwds={
               'loc': 'upper right',
               'bbox_to_anchor':(1,1)
            }                  # this puts the legend to the side
) 
```
